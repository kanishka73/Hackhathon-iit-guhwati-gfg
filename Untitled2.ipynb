```python
data1=training_features
data2=training_lables
data3=test_features
```

```python
# Identify categorical and numerical columns
categorical_features = ['age_group', 'education', 'race', 'sex', 'income_poverty',
                        'marital_status', 'rent_or_own', 'employment_status',
                        'hhs_geo_region', 'census_msa', 'employment_industry',
                        'employment_occupation']
numerical_features = [col for col in data1.columns if col not in categorical_features + ['respondent_id']]

from sklearn.impute import SimpleImputer


# Impute categorical features with the mode
categorical_imputer = SimpleImputer(strategy='most_frequent')
data1[categorical_features] = categorical_imputer.fit_transform(data1[categorical_features])
data3[categorical_features] = categorical_imputer.transform(data3[categorical_features])

# Impute numerical features with the median
numerical_imputer = SimpleImputer(strategy='median')
data1[numerical_features] = numerical_imputer.fit_transform(data1[numerical_features])
data3[numerical_features] = numerical_imputer.transform(data3[numerical_features])

```

```python
from sklearn.preprocessing import OneHotEncoder
# Create an instance of OneHotEncoder
encoder = OneHotEncoder()

# Fit and transform the training data
data1_encoded = encoder.fit_transform(data1[categorical_features])

# Transform the test data
data3_encoded = encoder.transform(data3[categorical_features])
```

```python
# Convert the sparse matrix to a Pandas DataFrame
from sklearn.preprocessing import  StandardScaler

# Use a different variable name to avoid overwriting
data1_encoded_sparse_df = pd.DataFrame.sparse.from_spmatrix(data1_encoded)
data3_encoded_df = pd.DataFrame.sparse.from_spmatrix(data3_encoded)

# Drop the original categorical columns and concatenate the encoded columns
data1 = data1.drop(columns=categorical_features + ['respondent_id'])
# Use a temporary variable for the concatenated DataFrame
data1_concat = pd.concat([data1.reset_index(drop=True), data1_encoded_sparse_df.reset_index(drop=True)], axis=1)

data3 = data3.drop(columns=categorical_features + ['respondent_id'])
data3 = pd.concat([data3.reset_index(drop=True), data3_encoded_df.reset_index(drop=True)], axis=1)

# Standardize the numerical features
scaler = StandardScaler()
numerical_features = data1.select_dtypes(include=['int64', 'float64']).columns

# Use the temporary variable for scaling
data1_concat[numerical_features] = scaler.fit_transform(data1_concat[numerical_features])
data3[numerical_features] = scaler.transform(data3[numerical_features])

# Now you can assign the scaled DataFrame back to data1_encoded if needed
data1_encoded = data1_concat
```

```python
import pandas as pd
from scipy.sparse import csr_matrix

# Assuming data1_encoded is the processed training data and data3 is the test data
# No need to convert to dense if they are already dense
# data1_encoded = data1_encoded.sparse.to_dense()  # Remove this line
# data3 = data3.sparse.to_dense()  # Remove this line

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

# Splitting the data into features and target variables
X = data1_encoded
y_xyz = data2['xyz_vaccine']
y_seasonal = data2['seasonal_vaccine']

# Splitting the data into training and validation sets for xyz_vaccine
X_train_xyz, X_val_xyz, y_train_xyz, y_val_xyz = train_test_split(X, y_xyz, test_size=0.2, random_state=42)

# Splitting the data into training and validation sets for seasonal_vaccine
X_train_seasonal, X_val_seasonal, y_train_seasonal, y_val_seasonal = train_test_split(X, y_seasonal, test_size=0.2, random_state=42)

# Convert all column names to strings
X_train_xyz.columns = X_train_xyz.columns.astype(str)
X_val_xyz.columns = X_val_xyz.columns.astype(str)


# Initialize the RandomForestClassifier for xyz_vaccine
model_xyz = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model
model_xyz.fit(X_train_xyz, y_train_xyz)

# Make predictions on the validation set
y_val_pred_proba_xyz = model_xyz.predict_proba(X_val_xyz)[:, 1]

# Calculate ROC AUC score for xyz_vaccine
xyz_vaccine_auc = roc_auc_score(y_val_xyz, y_val_pred_proba_xyz)

# Initialize the RandomForestClassifier for seasonal_vaccine
model_seasonal = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model
model_seasonal.fit(X_train_seasonal, y_train_seasonal)

# Make predictions on the validation set
y_val_pred_proba_seasonal = model_seasonal.predict_proba(X_val_seasonal)[:, 1]

# Calculate ROC AUC score for seasonal_vaccine
seasonal_vaccine_auc = roc_auc_score(y_val_seasonal, y_val_pred_proba_seasonal)

# Compute the mean ROC AUC score
mean_auc = (xyz_vaccine_auc + seasonal_vaccine_auc) / 2

print(f'ROC AUC for xyz_vaccine: {xyz_vaccine_auc:.4f}')
print(f'ROC AUC for seasonal_vaccine: {seasonal_vaccine_auc:.4f}')
print(f'Mean ROC AUC: {mean_auc:.4f}')
```
